```python
import requests
from bs4 import BeautifulSoup

```

```python
url = "https://www.goeduhub.com/"
response = requests.get(url)
response
htmlcontent = response.content
Soup =  BeautifulSoup(htmlcontent,"html.parser")
Soup.prettify
```

```python
Soup.findAll("title")
```

```python
print(Soup.findAll("div"))
```

```python
print(Soup.findAll("text"))
```

```python
one_tag = Soup.findAll('a')[4]
print(one_tag)
```

```python
anchor = Soup.findAll('a')
all_link = set()
for link in anchor:
    if (link.get('href')!='#'):
        linkprint ="https:www.goeduhub.com/"+link.get('href')
        all_link.add(link)
        print(linkprint)        
        
```
